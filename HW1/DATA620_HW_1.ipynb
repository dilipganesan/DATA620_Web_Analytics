```python
import networkx as nx
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
edges = {"Jane" : {"Ike"}, 
         "Ike" : {"Jane", "Heather"}, 
         "Heather": {"Ike", "Fernando", "Garth"},
         "Fernando": {"Heather", "Garth", "Diane", "Andre", "Carol"},
         "Garth": {"Heather", "Fernando", "Diane", "Beverly", "Ed"},
         "Diane": {"Fernando", "Garth", "Carol", "Andre", "Beverly", "Ed"},
         "Carol": {"Fernando", "Diane", "Andre"},
         "Andre": {"Fernando", "Diane", "Carol", "Beverly"},
         "Beverly": {"Andre", "Diane", "Garth", "Ed"},
         "Ed": {"Beverly", "Diane", "Garth"}
        }
```

```python
krachdartkite = nx.from_dict_of_lists(edges)
nx.number_of_nodes(krachdartkite)
nx.number_of_edges(krachdartkite)
```

```python
nx.draw(krachdartkite, with_labels=True)
```
